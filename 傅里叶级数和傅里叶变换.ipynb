# 傅里叶变换

傅里叶变换公式

时域函数到频域函数
$$F(\omega_x) = \int_{-\infty}^{+\infty}f(t)e^{-i\omega_x t}dt$$

频域函数到时域函数
$$f(t) = \frac{1}{2\pi}\int_{-\infty}^{+\infty}F(\omega_x)\cdot e^{i\omega_x t}d\omega$$

傅里叶级数公式
$$f(t) = \frac{a_0}{2} + \sum_{n-1}^\infty\left[a_n\cos(n\omega t + b_n\sin(n\omega t))\right]$$
其中
$$a_n = \frac{2}{T}\int_{t_0}^{t_0+T}f(t)\cos(n\omega t)dt$$
$$a_n = \frac{2}{T}\int_{t_0}^{t_0+T}f(t)\sin(n\omega t)dt$$

利用三角函数正交性，即可推导得到

参考自 https://zhuanlan.zhihu.com/p/41455378
周期函数是客观世界中周期运动的数学表达，如物体挂在弹簧上作简谐振动、单摆振动、无线电电子振荡器的电子振荡等，大多可以表述为：

$$ f(t) = A\sin(\omega t + \psi)$$


```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
#%matplotlib notebook
t = np.arange(-2*np.pi, 2*np.pi, 0.01)
psi = 0.0
omega = 1
A = 5
y = A*np.sin(omega*t+psi)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(t, y)
ax.set_title("sine wave")
ax.set_xlabel("angle")
ax.set_ylabel("sine")
plt.show()
```

思考，可否利用一系列三角函数$A_n\sin(n\omega t + \psi_n)$ 之和来表示较复杂的周期函数$f(t)$，公式如下
$$f(t) = A_0 + \sum_{n=1}^\infty(n\omega t + \psi_n)$$
傅里叶级数中，需要指定一个初始频率$\omega_0$，之后的频率必须是该频率的整数倍

**三角函数的正交性**

$$\int_{-\pi}^\pi{\cos nxdx} = 0  \qquad\qquad (n=1,2,3,\dots)$$
$$\int_{-\pi}^\pi{\sin nxdx} = 0  \qquad\qquad (n=1,2,3,\dots)$$
$$\int_{-\pi}^\pi{\sin kx \cdot\cos nxdx} = 0  \qquad\qquad (k,n=1,2,3,\dots)$$
$$\int_{-\pi}^\pi{\cos kx \cdot\cos nxdx} = 0  \qquad\qquad (k,n=1,2,3,\dots; k \neq n)$$
$$\int_{-\pi}^\pi{\sin kx \cdot\sin nxdx} = 0  \qquad\qquad (k,n=1,2,3,\dots; k \neq n)$$
