```python
import pandas
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib inline

# 幂函数
x = np.arange(0, 30, 0.1)
miu = 2

y = x**miu

figure, axe = plt.subplots(1, 1)
axe.plot(x, y)
axe.set_xlabel("X")
axe.set_ylabel("Y")
#axe.set_title("幂函数")
plt.show()
```

```python
# 指数函数
x = np.arange(0, 10, 0.1)
a = 2

y = a**x

figure, axe = plt.subplots(1, 1)
axe.plot(x, y)
axe.set_xlabel("X")
axe.set_ylabel("Y")
plt.show()
```

```python
# 对数函数
x = np.arange(0.001, 100, 1)
a = 2

#y = np.log(x)/np.log(a)
y = np.log(x)

figure, axe = plt.subplots(1, 1)
axe.plot(x, y)
axe.set_xlabel("X")
axe.set_ylabel("Y")
plt.show()
```

```python
# tanh函数
x = np.arange(-2, 2, 0.01)
y = (math.e**x - math.e**(-x))/(math.e**x + math.e**(-x))

figure, axe = plt.subplots(1, 1)
axe.plot(x, y)
axe.set_xlabel("X")
axe.set_ylabel
plt.show()
```
